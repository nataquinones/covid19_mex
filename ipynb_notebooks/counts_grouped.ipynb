# Cuentas y cuentas acumuladas

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
import datetime as datetime
```

```python
df = pd.read_csv('../datos/tablas_procesadas/tabla_completa.tsv',
                 sep='\t',
                 parse_dates=['fecha_inicio_sintomas', 'fecha_llegada_mx', 'fecha_caso_nuevo'])
df['fecha_caso_nuevo'] = df['fecha_caso_nuevo'].replace(pd.NaT, datetime.datetime(2020,3,15))
df['dummy_count'] = len(df) * [1]
df = df[['estado', 'procedencia', 'fecha_caso_nuevo', 'dummy_count']]
```

```python
df
```

## Cuenta por: estados, fechas, procedencia

```python
df2 = df.groupby(['estado', 'fecha_caso_nuevo', 'procedencia'])['dummy_count'].count().reset_index()
```

```python
df2.head()
```

## Cuenta acumulada por estados y fechas

```python
df2_cum = df.groupby(['estado', 'fecha_caso_nuevo', 'procedencia']).sum().groupby(level=[2]).cumsum().reset_index()
```

```python
df2_cum.head()
```

## Cuenta por: estados y fechas

```python
df3 = df2.groupby(['estado', 'fecha_caso_nuevo']).sum().reset_index()
```

```python
df3.head()
```

## Cuenta acumulada por estados y fechas

```python
df3_cum = df.groupby(['estado', 'fecha_caso_nuevo']).sum().groupby(level=[0]).cumsum().reset_index()
```
